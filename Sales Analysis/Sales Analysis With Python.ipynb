# 1. WHAT WAS THE BEST MONTH FOR SALE, HOW MUCH WE EARNED AT THAT MONTH?
* first check out all the columns. Their data types.
* TO CALCULATE SALES, WE HAVE TO FIRST CREATE A COLUMN FOR SALES.

```python
df.dtypes
```

```python
df['quantity_ordered'] = pd.to_numeric(df['quantity_ordered'])
df['price_each'] = pd.to_numeric(df['price_each'])
```

```python
df.dtypes
```

```python
df['sales'] = df['quantity_ordered'] * df['price_each']
```

```python
df.head()
```

### now find the month

```python
df.groupby('month').sum()
```

* we see December has the highest sales [12]
* let's plot it.

```python
highest_earned_month = df.groupby('month').sum()['sales']
```

```python
quantity_ordered_month = df.groupby('month').sum()['quantity_ordered']
```

```python
months = range(1,13)
```

```python
plt.figure(figsize = (12,6))
plt.style.use('fivethirtyeight')
plt.title('Sales Per Month')
width = 0.45
plt.bar(months,highest_earned_month, width = width, color = '#F08080', label='Sales')
plt.legend()
plt.xticks(months)
plt.xlabel('Months')
plt.ylabel('Sales per month in USD')
plt.show()
```

```python
plt.figure(figsize = (12,6))
plt.style.use('Solarize_Light2')
plt.title('Sales VS Month')
plt.plot(months,highest_earned_month,color = '#F08080', label='Sales')
plt.legend()
plt.xticks(months)
plt.xlabel('Months')
plt.ylabel('Sales per month in USD')
plt.show()
```

# 4.  WHAT PRODUCTS ARE MOST OFTEN SOLD TOGETHER?
* Firstly, check the duplicate order_id. duplicate order id shows, people bought 2 or more things at the same time.
* then we have to join the prodcuts of same order id, into a column.

```python
df['product'].unique()
```

```python
# we just want to see the duplicate order_id's
df3 = df[df['order_id'].duplicated(keep = False)]
```

```python
df3.head(30)
```

```python
df3['together_sold_products'] = df3.groupby('order_id')['product'].transform(lambda x:','.join(x))
```

```python
# now we have joined all the products which are bought together
df3.head()
```

```python
df3 = df3[['order_id','together_sold_products']].drop_duplicates()
```

```python
df3.head()
```

```python
from itertools import combinations
from collections import Counter
```

```python
count = Counter()
```

```python
for row in df3['together_sold_products']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))
    
    
count.most_common(50)
```
