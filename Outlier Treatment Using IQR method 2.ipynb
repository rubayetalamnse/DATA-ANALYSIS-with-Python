```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
matplotlib.rc("figure",figsize = (12,6))
plt.style.use("ggplot")
```

```python
df = pd.read_csv('weight-height.csv')
```

```python
df.shape
```

```python
df.describe()
```

```python
q1 = df.Height.quantile(.25)
```

```python
q3 = df.Height.quantile(.75)
```

```python
iqr = q3 - q1
```

```python
iqr
```

```python
lower_limit = q1 - 1.5*iqr
```

```python
upper_limit = q3 + 1.5*iqr
```

```python
lower_limit
```

```python
upper_limit
```

```python
#everything beyond lower and upperlimit is outlier:
df[df.Height<lower_limit]
```

```python
df[df.Height>upper_limit]
```

```python
df_no_outlier = df[(df.Height<upper_limit)&(df.Height>lower_limit)]
```

```python
df_no_outlier.shape
```

# For Weight:

```python
df_no_outlier.Weight.describe()
```

```python
q1_weight = df_no_outlier.Weight.quantile(.25)
```

```python
q3_weight = df_no_outlier.Weight.quantile(.75)
```

```python
IQR = q3_weight-q1_weight
```

```python
IQR
```

```python
lower_limit_weight = q1_weight - 1.5*IQR
lower_limit_weight
```

```python
upper_limit_weight = q3_weight + 1.5* IQR
upper_limit_weight
```

```python
df_no_outlier[df_no_outlier.Weight<lower_limit_weight]
```

```python
df_no_outlier[df_no_outlier.Weight>upper_limit_weight]
```
