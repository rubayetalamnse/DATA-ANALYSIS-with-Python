# ALL THE FUNCTIONS DISCUSSED:
- df.dropna() #drop/remove all the NaN values
- df.dropna(how='any') #drop a row if there's 1 NaN value
- df.dropna(how='all') # drop a row if all the values are NaN
- df.isnull() # true/false = output
- df.isnull().sum() #how many missing values in each column
- df.isnull().sum().sum() #total missing values in all column
- df.fillna(value = 0)
- df.fillna(method = 'pad') #filling the missing values with the previous value (row-wise)
- df.fillna(method = 'pad', axis = 1)#filling the missing values with the previous value (column-wise)
- df.fillna(method = 'bfill') #filling the missing values with the following value (row-wise)
- df.fillna(method = 'bfill', axis = 1)#filling the missing values with the following value (column-wise)
- df.fillna({'column1':'Not Found', 'column2':'Not Available'}) #filling the missing values with certain words for a certain column
- df.fillna(value=df['numerical_column'].mean()) #filling the missing values of a numerical column with the mean of that column
- df.replace(to_replace = np.nan, value = 2022) #replacing NaN values with any number
- df['numerical_column'] = df['numerical_column'].interpolate(method = 'linear') #linear method replaces the null value assuming a linearly spaced column. If the upper value is 0, middle is NaN, the lower value is 1. linear method will replace the null value with 0.5
- df['numerical_column'] = df['numerical_column'].interpolate(method = 'linear',limit_direction = 'forward') #using forward only 1 NaN value will be replaced.
